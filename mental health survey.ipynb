### Project Summary:

Mental health has always been the central topic around workplace culture and wellness conversation. It is increasingly recognized that employees' mental health is a crucial determinant in their overall health. Poor mental health at the workplace can contribute to a range of physical illnesses. Mental health can also affect their personal and professional lives. Organizations or employers need to recognize how mental wellness can impact their employees' productivity and performance.Though mental health issues in the workplace are important and need to be discussed, mental health is still heavily stigmatized. Many people don't feel comfortable discussing their mental health problems or admit that they are dealing with them for fear of being judged by others. 
This dataset is from a 2014 survey that measures attitudes towards mental health and frequency of mental health disorders in the tech workplace. The survey data contains responses to 27 questions by 1259 participants. Each question is a column with categorical values and each row represents response from a unique person.

### Problem Statement:

The objective in this project is to analyze the attitude of tech workers towards mental health and the main predictors of mental health illness in the U.S.(primarily) tech workplace. We explore and analyze the dataset "Mental Health in Tech workplace survey", with questions pertaining to how mental health is perceived at tech workplaces by employees and their employers.


### Research Questions

What are the provisions at tech workplaces to address mental health issues in employees? 
Do employees feel that the tech industry supports mental health and wellbeing? 
Are employers providing healthcare benefits for mental health issues? 
Are employees aware of their healthcare benefits and resources?
How do employers weigh mental health issues at the same level as to other medical issues?
Are there any leave policies for those suffering from mental health issues?

### Bivariate Analysis


Bivariate analysis involves analysing two variables by displaying the frequency distribution of one variable within the categories of another.
We have taken treatment as our target variable as our main objective is poeple who are affected by their mental health.

### Conclusion

The above exploration validates the initial hypothesis of growing concerns regarding mental health issues in the tech industry. The findings and insights of this research are limited to the participants who responded to the survey and may not be a generalization of the entire tech industry and workplaces.
Majority of the participants (84%) feel that the support to mental health in workplaces in tech is not adequate. While many participants shared that the importance to physical health by their employers was high, the relative importance to mental health has been much lower. The leave policies seem to be in favor of employees, but employers should take more efforts to make employees aware and comfortable of seeking time off for mental wellbeing. A large number of participants (50%) do not have medical healthcare coverage and benefits for mental health issues. A vast majority (70%) do not have access to open discussions and resources to learn more about mental health issues.

Steps that may be implemented at workplaces to ensure well being of the workers:

A study of this kind should be conducted at all workplaces to assess how employees feel about mental health disorders and what employers can do to address their concerns. Medical healthcare plans should be more inclusive allowing employees to seek medical attention for these disorders. Additionally, to overcome the taboo associated with these issues, employers should facilitate more open discussions and provide resources to learn more.
